```python

from PyQt5 import QtWidgets,QtCore,QtGui
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
import sys
import psutil

class MyWindow(QMainWindow):
    def __init__(self):
        super(MyWindow,self).__init__()
        self.setGeometry(200,200,768,1024) 
        self.setWindowTitle("System Monitor")
        self.initUI()
        
    def initUI(self):
        self.label = QtWidgets.QLabel(self)
        shdw = QGraphicsDropShadowEffect()
        shdw.setBlurRadius(15)
        self.label.setText("SYSTEM MONITOR")
        self.label.setFont(QFont("Impact",57))
        self.label.move(550,550)
        self.label.adjustSize()
        self.label.setStyleSheet("color: #11048D")
        self.label.setGraphicsEffect(shdw)
        self.label.setVisible(True)
            
            
        shadow = QGraphicsDropShadowEffect()
        shadow.setBlurRadius(15)
        self.button1 = QtWidgets.QPushButton(self)
        self.button1.setText("START")
        self.button1.setFont(QFont('Dubai Medium',16))
        self.button1.resize(500,75)
        self.button1.move(750,800)
        self.button1.setStyleSheet("background-color: #E3DEDE")
        self.button1.setGraphicsEffect(shadow)
        self.button1.clicked.connect(self.cpugen)
        
        shadw = QGraphicsDropShadowEffect()
        shadw.setBlurRadius(15)
        self.button2 = QtWidgets.QPushButton(self)
        self.button2.setText("EXIT")
        self.button2.move(750,950)
        self.button2.setFont(QFont('Dubai Medium',16))
        self.button2.resize(500,75)
        self.button2.clicked.connect(self.click1)
        self.button2.setStyleSheet("background-color: #E3DEDE")
        self.button2.setGraphicsEffect(shadw)
        
    def cpugen(self):
        self.label.hide()
        self.button1.hide()
        self.button2.hide()
        app.setStyleSheet(stylesheet1)
        
        shadw = QGraphicsDropShadowEffect()
        shadw.setBlurRadius(15)
        statistics = {}
        physical_and_logical_cpu_count = psutil.cpu_count()
        statistics['physical_and_logical_cpu_count'] = physical_and_logical_cpu_count
        cpu_load = [x / psutil.cpu_count() * 100 for x in psutil.getloadavg()][-1]
        statistics['cpu_load'] = cpu_load
        
        self.label1 = QLabel("CPU Usage:{}".format(statistics), self)
        self.label1.setFont(QFont("Candara",26))
        self.label1.move(150,250)
        self.label1.adjustSize()
        self.label1.setStyleSheet("color: #FFFFFF")
        self.label1.setGraphicsEffect(shadw)
        self.label1.setVisible(True)
    
        statram = {}
        statram = psutil.virtual_memory()
        self.label2 = QLabel("Memory Load:{}".format(statram), self)
        self.label2.setFont(QFont("Candara",26))
        self.label2.move(150,400)
        self.label2.adjustSize()
        self.label2.setStyleSheet("color: #FFFFFF")
        self.label2.setGraphicsEffect(shadw)
        self.label2.setVisible(True)
  
        statdisk = {}
        statdisk = psutil.disk_usage('/')
        
        self.label3 = QLabel("Disk Usage:{}".format(statdisk), self)
        self.label3.setFont(QFont("Candara",26))
        self.label3.move(150,550)
        self.label3.adjustSize()
        self.label3.setStyleSheet("color: #FFFFFF")
        self.label3.setGraphicsEffect(shadw)
        self.label3.setVisible(True)
     
        statnet={}
        statnet=psutil.net_io_counters()
        self.label4 = QLabel("Net Usage:{}".format(statnet), self)
        self.label4.setFont(QFont("Candara",26))
        self.label4.move(150,700)
        self.label4.adjustSize()
        self.label4.setStyleSheet("color: #FFFFFF")
        self.label4.setGraphicsEffect(shadw)
        self.label4.setVisible(True)
      
        self.setStyleSheet("border : 3px solid white;")
        
        self.button3 = QtWidgets.QPushButton(self)
        self.button3.setText("REFRESH")
        self.button3.setFont(QFont('Dubai Medium',16))
        self.button3.move(650,950)
        self.button3.resize(500,75)
        self.button3.setStyleSheet("background-color: #E3DEDE")
        self.button3.setGraphicsEffect(shadw)
        self.button3.setVisible(True)
        self.button3.clicked.connect(self.rfersh)
        
        self.button4 = QtWidgets.QPushButton(self)
        self.button4.setText("EXIT")
        self.button4.setFont(QFont('Dubai Medium',16))
        self.button4.move(1200,950)
        self.button4.resize(500,75)
        self.button4.setStyleSheet("background-color: #E3DEDE")
        self.button4.setGraphicsEffect(shadw)
        self.button4.setVisible(True)
        self.button4.clicked.connect(self.click1)
        
    def rfersh(self):
        self.label1.clear()
        self.label2.clear()
        self.label3.clear()
        self.label4.clear()
        self.cpugen()
    def click1(self):
        self.close()
      
    def update(self):
        self.label.adjustSize()
        
        
        
stylesheet = """
    MyWindow {
        background-image: url("C:/Users/rhyth/Downloads/technology-1650304985164-2957.jpg"); 
        background-repeat: no-repeat; 
        background-position: center;
        background-size: stretch;
    }
"""
stylesheet1 = """
    MyWindow { background-color: blue;}
    """
app = QApplication(sys.argv)

def window():
    
    app.setStyleSheet(stylesheet) 
    win = MyWindow()
    win.resize(1980, 1080)
    win.show()
    sys.exit(app.exec_())

window()
```
